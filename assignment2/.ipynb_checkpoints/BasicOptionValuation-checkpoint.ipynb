# Part 1: Basic Option Valuation

```python
class Properties():
    
    def __init__(self):
        
        self.T = 1
        self.sigma = 0.2
        self.r = 0.06
        self.s0 = 100
        self.K = 99
        self.N = 365
        self.dt = self.T/self.N
        self.M = 1000
        
```

```python
import math
import numpy as np

def ST(props):
    return props.s0 * math.exp((props.r-0.5*props.sigma**2)*props.T+props.sigma*props.T**0.5*np.random.normal())

def option_price(props):
    
    thesum = 0
    
    for i in range(props.M):
        thesum += max(ST(props)-props.K,0)
        
    average = thesum / props.M
    price = math.exp(-props.r*props.T) * average
    return(price)

proper = Properties()
for i in range(50):
    print(option_price(proper))
```

```python
proper = Properties()

values = []

for i in range(50000,200000,25000):
    
    
    for k in range(50000,200000,25000):
        proper.M = k

        values.append([])

        for j in range(i):
            values[-1].append(option_price(proper))
            
        print(i,k)
    
    save_values(values)
    values = []
```

```python
import csv

def save_values(data):
    
    filename = 'part1data/putvalues.csv'
    
    with open(filename, 'a', newline = '') as csvfile:
            writer = csv.writer(csvfile, delimiter=',', quotechar='"')
        
            writer.writerow(data)
                
```

```python
import csv
import matplotlib.pyplot as plt

props = Properties()

def read_data_convergence(M,N):
    
    thesum = 0
    prices = []

    with open("part1data/thedata.csv", "r") as csvfile:

        reader = csv.reader(csvfile, delimiter=',', quotechar='"', quoting=csv.QUOTE_NONNUMERIC)
        counter = 0

        # For each reflection with this mua, it is checked if it is in
        # right range for r
        for row in reader:
            counter+= 1
            thesum += row[0]
            
            if counter == M:
                print("progress")
                average = thesum / M
                price = math.exp(-props.r*props.T) * average
                prices.append(price)
                thesum = 0
                counter = 0
                if len(prices)==N:
                    print("lack of progress")
                    break

    save_values(prices)
        
for i in [100,1000,10000]:
    for j in [1000,10000,100000]:
        
        read_data_convergence(j,i)
```
